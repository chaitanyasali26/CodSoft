# Lets Train our model using linear regression 

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train, y_test = train_test_split(advertising[['TV']],advertising[['Sales']],test_size = 0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
model.intercept_
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,6.948 + 0.054*X_test , 'r')
plt.show()
```

Conclusion:- We can say that above model is able to predict Sales
